```python
#birds['westBoundingCoordinate'].astype('float')
birds['eastBoundingCoordinate'] = birds['eastBoundingCoordinate'].str.replace(",", ".").astype(float)
```

```python
birds['northBoundingCoordinate'] = birds['northBoundingCoordinate'].str.replace(",", ".").astype(float)
```

```python
# Define a custom function to combine columns
def mgrs_compute(row):
    return mgrs.MGRS().toMGRS(row['northBoundingCoordinate'], row['eastBoundingCoordinate'])
```

```python
def mgrs_compute2(row):
    return mgrs.MGRS().toMGRS(row['northBoundingCoordinate'].str.replace(",", ".").astype(float), row['eastBoundingCoordinate'].str.replace(",", ".").astype(float))
```

```python
# Apply the custom function to create a new column 'MGRS'

#birds['MGRS'] = birds.apply(mgrs_compute, axis=1)
#ValueError: could not convert string to float: '47,84122'  previously I converted string to float #birds['northBoundingCoordinate'] = birds['northBoundingCoordinate'].str.replace(",", ".").astype(float)

birds['MGRS'] = birds.apply(mgrs_compute, axis=1)
```
